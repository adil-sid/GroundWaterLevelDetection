```python
import xarray as xr
import pandas as pd

# open your NetCDF file
ds = xr.open_dataset(r"D:\GWL Detection\GroundWaterLevelDetection\data\Copernicus_Temperature\Delhi_2m_temperature202001.nc")

# check variables
print(ds)

# select your temperature variable
var = "t2m"   # change if needed
da = ds[var]

# convert from Kelvin → Celsius
da = da - 273.15

# compute mean over valid_time
monthly_mean = da.mean(dim="valid_time")

# convert to DataFrame
df_mean = monthly_mean.to_dataframe(name="tmean_C").reset_index()

# save to CSV
df_mean.to_csv("temperature_monthly_mean_by_cell.csv", index=False)

print("✅ Saved temperature_monthly_mean_by_cell.csv")
```

```python
import xarray as xr
import pandas as pd
import glob
import os

# folder with nc files
files = glob.glob(r"D:\GWL Detection\GroundWaterLevelDetection\data\Copernicus_Temperature\*.nc")

all_months = []

for file in files:
    print("Processing:", file)
    ds = xr.open_dataset(file)
    
    da = ds["t2m"] - 273.15
    monthly_mean = da.mean(dim="valid_time")
    df_mean = monthly_mean.to_dataframe(name="tmean_C").reset_index()
    
    # ✅ get last 6 digits = YYYYMM
    fname = os.path.basename(file)
    yyyymm = "".join([c for c in fname if c.isdigit()])[-6:]
    month_date = pd.to_datetime(yyyymm, format="%Y%m")
    df_mean["month"] = month_date.strftime("%Y-%m")
    
    all_months.append(df_mean)


# combine all months
df_all = pd.concat(all_months, ignore_index=True)

# save
df_all.to_csv("temperature_monthly_mean_by_cell_all.csv", index=False)

print("✅ Saved temperature_monthly_mean_by_cell_all.csv")
```

```python
import xarray as xr
import pandas as pd
import glob
import os

# path to your Copernicus netCDF files
files = sorted(glob.glob(r"D:\GWL Detection\GroundWaterLevelDetection\data\Copernicus_Temperature\*.nc"))

all_months = []

for file in files:
    print("📂 Processing:", file)
    ds = xr.open_dataset(file)

    # Temperature variable (Kelvin → Celsius)
    da = ds["t2m"] - 273.15

    # Take monthly mean across all time steps
    monthly_mean = da.mean(dim="valid_time")

    # Convert to DataFrame
    df_mean = monthly_mean.to_dataframe(name="tmean_C").reset_index()

    # Round to 2 decimal places
    df_mean["tmean_C"] = df_mean["tmean_C"].round(2)

    # Extract YYYYMM from filename (last 6 digits only)
    fname = os.path.basename(file)
    yyyymm = "".join([c for c in fname if c.isdigit()])[-6:]

    # Convert to datetime and format
    month_date = pd.to_datetime(yyyymm, format="%Y%m")
    df_mean["month"] = month_date.strftime("%Y-%m")

    all_months.append(df_mean)

# Combine all months
df_all = pd.concat(all_months, ignore_index=True)

# Save to CSV
df_all.to_csv("temperature_monthly_mean_by_cell_all.csv", index=False)
print("✅ Saved: temperature_monthly_mean_by_cell_all.csv")
```
