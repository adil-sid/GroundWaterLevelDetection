```python
import xarray as xr
import pandas as pd
import glob
import os

# folder with nc files
files = glob.glob(r"D:\GWL Detection\GroundWaterLevelDetection\data\Copernicus_Temperature\*.nc")

all_months = []

for file in files:
    print("Processing:", file)
    ds = xr.open_dataset(file)
    
    da = ds["t2m"] - 273.15
    monthly_mean = da.mean(dim="valid_time")
    df_mean = monthly_mean.to_dataframe(name="tmean_C").reset_index()
    
    # ✅ get last 6 digits = YYYYMM
    fname = os.path.basename(file)
    yyyymm = "".join([c for c in fname if c.isdigit()])[-6:]
    month_date = pd.to_datetime(yyyymm, format="%Y%m")
    df_mean["month"] = month_date.strftime("%Y-%m")
    
    all_months.append(df_mean)


# combine all months
df_all = pd.concat(all_months, ignore_index=True)

# save
df_all.to_csv("temperature_monthly_mean_by_cell_all.csv", index=False)

print("✅ Saved temperature_monthly_mean_by_cell_all.csv")
```

```python
# Load groundwater CSV
df = pd.read_csv(r"D:\GWL Detection\GroundWaterLevelDetection\data\filtered_delhi_gwl.csv")

# Ensure datetime
df["dataTime"] = pd.to_datetime(df["dataTime"])

# Extract month
df["month"] = df["dataTime"].dt.to_period("M").astype(str)

# Group by lat, lon, month
df_mean = (
    df.groupby(["latitude", "longitude", "month"], as_index=False)
    .agg({
        "dataValue": "mean",
        "wellDepth": "first"   # keep depth (same for that well)
    })
)

# Round values
df_mean["dataValue"] = df_mean["dataValue"].round(2)
df_mean["wellDepth"] = df_mean["wellDepth"].round(2)

# Save
df_mean.to_csv("gwl_monthly_mean.csv", index=False)
print("✅ Saved: gwl_monthly_mean.csv")
```

```python
import pandas as pd

def merge_data(rainfall_file, gwl_file, output_file):
    """
    Merges rainfall and groundwater level data based on latitude, longitude, and month.

    Args:
        rainfall_file (str): Path to the rainfall data CSV file.
        gwl_file (str): Path to the groundwater level data CSV file.
        output_file (str): Path to save the merged CSV file.
    """
    try:
        # Load the two datasets
        print(f"Reading rainfall data from '{rainfall_file}'...")
        df_rain = pd.read_csv(rainfall_file)

        print(f"Reading groundwater level data from '{gwl_file}'...")
        df_gwl = pd.read_csv(gwl_file)

        print("\nRainfall data sample:")
        print(df_rain.head())

        print("\nGroundwater level data sample:")
        print(df_gwl.head())

        # --- Data Merging ---
        # We will merge the two dataframes based on the columns they share:
        # 'latitude', 'longitude', and 'month'.
        # An 'inner' merge ensures that we only keep rows where the location
        # and month exist in *both* files.
        merge_columns = ['latitude', 'longitude', 'month']
        print(f"\nMerging the two datasets on: {', '.join(merge_columns)}...")
        
        df_merged = pd.merge(df_rain, df_gwl, on=merge_columns, how='inner')

        print("\nMerged data sample:")
        print(df_merged.head())

        # Save the merged dataframe to a new CSV file
        df_merged.to_csv(output_file, index=False)
        print(f"\n✅ Successfully merged data and saved to '{output_file}'.")
        print(f"The new file has {len(df_merged)} rows.")

    except FileNotFoundError as e:
        print(f"❌ Error: File not found. Please check the file path: {e.filename}")
    except KeyError as e:
        print(f"❌ Error: A required column was not found in one of the files: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == '__main__':
    # Define the input and output file names
    rainfall_input = r'F:\AI_Projects\groundwater\GroundWaterLevelDetection\research\merged_rainfall_and_gwl.csv'
    gwl_input = r'F:\AI_Projects\groundwater\GroundWaterLevelDetection\research\temperature_monthly_final.csv'
    merged_output = 'merged_temp_and_mer.csv'

    # Run the merge function
    merge_data(rainfall_input, gwl_input, merged_output)
```
